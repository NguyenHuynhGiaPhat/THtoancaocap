a) 1/x  x  1/x-3

```python
from sympy.solvers.solvers import denoms 
eq = (1/x)*1/(x-3)
dd = denoms(eq)
print(dd)
```

b) f = (1+1/x)/x-11

```python
eq = (1+1/x)/(x-1)
from sympy.solvers.solvers import denoms
loai_tru = set()
for d in denoms(eq):
  for s in solve(d):
    loai_tru.add(s)
print(loai_tru)
```

Thực hành 13:


```python
from sympy import Symbol, Limit
t = Symbol('t')
St = 5*t**2 + 2*t +8
t1 = Symbol('t1')
delta_t = Symbol('delta_t')
St1 = St.subs({t: t1})
St1_delta = St.subs({t: t1 + delta_t})
Limit((St1_delta-St1)/delta_t, delta_t,0).doit()
```
